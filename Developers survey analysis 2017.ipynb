# 2.What are majority of Developers doing?

```python
ct = survey_data['DeveloperType'].value_counts().reset_index()
ct.head()
```

```python
#cleaning function that is cleaning the data
#here we are not dealing with null values because it will automatically not considered 

def clean_series(df,col_name):
    """
    inputs: 
    df: dataframe which you want to manipulate
    col_name: name of the column which you want to clean the data
    
    outputs:
    df2: Panda's dataframe with the unique element seperated and
         their count
    """
    temp = df[col_name]
    temp = temp.dropna().reset_index()  #dropna() is used here to drop rows with NaN VALUES IN THE DATAFRAME
    temp = temp[col_name].str.split(';')
    emp_list=[]
    for i in range(len(temp)):
        emp_list += temp[i]

    for i in range(len(emp_list)):
        emp_list[i] = emp_list[i].strip()

    emp_set = list(set(emp_list))    
    emp_set

    c = Counter(emp_list)
    print(c.items())
    df2 = pd.DataFrame({'DeveloperType':list(c.keys()),'Counts':list(c.values())})
    
    return df2
```

```python
dev = clean_series(survey_data,'DeveloperType').sort_values(by='Counts')
dev
```

```python
plt.figure(figsize=(20,6))
ax=sns.barplot(data = dev , x = 'DeveloperType',y = 'Counts',)
ax.set_xlabel('DeveloperType')
ax.set_xticklabels(dev['DeveloperType'], rotation='vertical', fontsize=10)
```

Web developers were at the top in terms of numbers , more than 25000 developers were web developers,

# 3. Which platform was mostly used by developers in 2017 and what are their future plans of upgrading ?

```python
def get_popular_given_column_order(df: pd.DataFrame, column: str) -> pd.DataFrame:
    """
    INPUT: 
        1. it takes data frame with data from stackoverflow 2018 survey
        2. column name which we use for analysis
    OUTPUT: 
        return the platform name with its usage count worldwide in a data frame
    """
    data = {}
    for value in df[column].values:
        if isinstance(value, float):
            continue
        for platform in value.split(";"):
            data[platform] = data.get(platform, 0) + 1
    
    df = pd.DataFrame(data.items(), columns=['Platform', 'Count']).rename({"Platform": column})
    df = df.sort_values('Count', ascending=False)
    return df
```

```python
df = get_popular_given_column_order(survey_data, 'HaveWorkedPlatform')
df.head()
```

```python
plt.figure(figsize=(25,10))
sns.barplot(data=df,x='Count', y='Platform')
```

Android and Windows Desktop are the most used platforms.

```python
ans_3 = get_popular_given_column_order(survey_data, 'WantWorkPlatform')
ans_3.head()
```

```python
plt.figure(figsize=(25,10))

sns.barplot(data=ans_3,x='Count', y='Platform')
```

Most of the people desire to continue with Android but a major portion of windows desktop users want to upgraded to android ,AWS and raspberry pi due to increase demand

# 4.Which country averagely pays more to it's developers ?

```python
ans_4=survey_data.groupby(['Country']).mean().reset_index()
a=ans_4[['Country','Salary']].sort_values(by='Salary')
b=a.dropna() #for not considering the the NaN values
b
```

```python
plt.figure(figsize=(50,20))
ax=sns.barplot(data = b , x = 'Country',y = 'Salary',)
ax.set_xlabel('Country')
ax.set_xticklabels(b['Country'], rotation='vertical', fontsize=10)
```

By looking at the visuals developers would be earning most in Virgin Islands (USA)
